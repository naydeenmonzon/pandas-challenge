Your responsibility is to aggregate the data to and showcase obvious trends in school performance.

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data
```

```python
student_data
```

```python
# Combine the data into a single dataset.  
complete_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
complete_data
```

## District Summary

Create a high level snapshot (in table form) of the district’s key metrics, including:
   * Total Schools
   * Total Students
   * Total Budget
   * Average Math Score
   * Average Reading Score
   * % Passing Math (The percentage of students that passed math.)
   * % Passing Reading (The percentage of students that passed reading.)
   * % Overall Passing (The percentage of students that passed math and reading.)

```python
# Check to see if there are any rows with missing data
complete_data.count()
```

```python
# Total Schools
school_count = school_data["school_name"].count()
# Total Students
student_count = student_data["student_name"].count()
# Total Budget
total_budget = school_data["budget"].sum()
# Average Math Score
average_math = (complete_data["math_score"].mean())
# Average Reading Score
average_reading = complete_data["reading_score"].mean()
# % Passing Math (The percentage of students that passed math.)
passing_math = ((complete_data["math_score"] >= 70).sum() / student_count) * 100
# % Passing Reading (The percentage of students that passed reading.)
passing_reading = ((complete_data["reading_score"] >= 70).sum() / student_count * 100)
# % Overall Passing (The percentage of students that passed math and reading.)
overall_passing = ("{:.2f} %".format((((complete_data["math_score"] >= 70) & (complete_data["reading_score"] >= 70)).sum() / student_count) * 100))
```

```python
# Now that we have all the series needed to complete the District_Summary df
District_Summary = pd.DataFrame({
    "Total Schools": school_count,
    "Total Students": ('{:,}'.format(student_count)),
    "Total Budget": ('${:,.2f}'.format(total_budget)),
    "Average Math Score": ('{:.2f}'.format(average_math)),
    "Average Reading Score": ('{:.2f}'.format(average_reading)),
    "% Passing Math": ("{:.2f} %".format(passing_math)),
    "% Passing Reading": ("{:.2f} %".format(passing_reading)),
    "% a Passing": [overall_passing]
})

District_Summary
```

## School Summary

Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math and reading.)
  
Create a dataframe to hold the above results

```python
#Since most of the metrics being asked is the average per school, let's get the mean value by school_name from the comple_data set.
#Include the type so it will be added to the data frame
school_df = complete_data.groupby(['school_name','type']).mean()
school_df
```

```python
# From the school_df we have the following series we need for our School_Summary_df:
    # School Name = "school_name"
    # School Type = "type"
    # Total Students = "size"
    # Total School Budget = "budget"    
    # Average Math Score = "math_score"
    # Average Reading Score = "reading_score"

    
# We need the following series to complete the School_Summary_df:
    # Per Student Budget
    # % Passing Math
    # % Passing Reading
    # % Overall Passing (The percentage of students that passed math and reading.)

    
# # Per Student Budget
per_student_budget = school_df.groupby(["school_name"]).mean()["budget"] / school_df["size"]


# Filter all the math, reading and overall score that meets the passing grade from the complete_data df
passing_math = complete_data[(complete_data["math_score"] >= 70)]
passing_reading = complete_data[(complete_data["reading_score"] >= 70)]
passing_overall = complete_data[(complete_data["reading_score"] >= 70) & (complete_data["math_score"] >= 70)]

# Use the filtered list to get the scores per school...
# % Passing Math
school_passing_math = passing_math.groupby(['school_name']).count()['student_name'] / school_df["size"] * 100
# % Passing Reading
school_passing_reading = passing_reading.groupby(['school_name']).count()['student_name'] / school_df["size"] * 100
# % Overall Passing (The percentage of students that passed math and reading.)
school_passing_overall = passing_overall.groupby(['school_name']).count()['student_name'] / school_df["size"] * 100


# Let's add the new series into school_df
school_df = school_df.join(pd.DataFrame(
    {
    "Per Student Budget": per_student_budget,
    "% Passing Math": school_passing_math,
    "% Passing Reading": school_passing_reading,
    "% Overall Passing": school_passing_overall
    }, index=school_df.index
))
school_df
```

```python
# Now let's make the final School_Summary by removing some columns we dont need and re-naming

school_df.drop(["Student ID"], axis=1)

School_Summary_df = school_df[["size", "budget", "Per Student Budget", "math_score", "reading_score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]
School_Summary_df
```
